## Summary/background

**Prompted by an online discussion of media focus and the racial demographics of violent crime, I produced a bar graph (using Seaborn) to show as clearly as possible which racial combinations (offender/victim) are above and which are below what would be expected based on the size of racial groups alone (i.e. assuming random victims).**

**Data taken from the National Crime Victimization Survey**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
races = ['White/White','Black/Black','Hispanic/Hispanic',' ','White/Black','Black/White',' ','White/Hispanic',
         'Hispanic/White',' ','Black/Hispanic','Hispanic/Black']
```

```python
multipliers = [1.25,4.33,1.83,1,0.15,1.33,1,0.37,0.64,1,0.85,0.35]
log_multipliers = np.log2(multipliers)
```

```python
x_labels = ['White / White','Black / Black','Hispanic / Hispanic',' ','White / Black','Black / White',' ','White / Hispanic',
         'Hispanic / White',' ','Black / Hispanic','Hispanic / Black']
y_values = [-2,-1,0,1,2]
y_labels = ['1/4x','1/2x','1x','2x','4x']
```

```python
data = pd.DataFrame({'races':races,'multipliers':multipliers,'log_multipliers':log_multipliers})
plt.figure(figsize=(12,6))
sns.set(style='darkgrid')
sns.barplot(data=data,x=data.index,y='log_multipliers',color='gray')
plt.title('Violent Crime by Race: Actual vs. Expected Share (Based on Population)')
plt.xlabel('Race (Offender/Victim)')
plt.ylabel('Actual vs. Expected Share')
plt.xticks(data.index,x_labels,rotation=55)
plt.yticks(y_values,y_labels)
plt.tick_params(bottom='on',direction='inout')
plt.axhline(0,lw=1,color='black')
```
