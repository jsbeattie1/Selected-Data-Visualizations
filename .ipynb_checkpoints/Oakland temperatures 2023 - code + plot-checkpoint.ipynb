## Summary/background

**The first half of 2023 was unseasonably cold in Oakland, CA. To show this, I recorded the difference between the actual high temperature for each day and the historical average for that date, and displayed the results on an annotated line graph (using Matplotlib).** 

**Temperature data taken from Weather Underground** 

```python
# Temperature differences added daily from most recent (July 19) to least recent (January 1)
temps = [-7,-7,-1,0,3,-2,-4,-8,-2,-2,-12,-10,-12,-9,-7,-8,-7,3,11,-1,-4,-12,-11,-11,-11,-9,-8,-6,-8,-4,-5,-6,-6,-3,-4,-5,-7,-5,-5,-8,-6,-4,-3,-5,-4,-4,0,-4,-6,-9,-6,-3,-11,-7,-6,-6,-6,-4,-1,-2,-7,-7,-6,-4,2,-3,3,10,
         -2,-5,-7,-4,-4,-5,-8,-8,-8,-9,-11,-9,-4,-6,-1,13,7,5,-1,-4,0,3,-2,-6,-9,-9,-8,-7,-8,-5,-6,-3,1,5,-3,-8,-6,-8,-9,
         -8,-7,-5,-6,-8,-13,-11,-4,-8,-6,-6,-6,-6,-8,-8,-6,3,1,-4,-4,-1,1,-8,-3,-6,-8,-9,-8,-8,-9,-10,-7,-7,-6,-8,-11,-12,
        -13,-12,-11,-14,3,0,-1,1,-8,-4,-8,-1,1,-5,-4,1,1,-4,-1,-4,-1,-4,-2,-4,-3,-4,-9,-5,-3,10,6,1,5,3,-2,-3,-5,-4,-4,-3,
        -3,1,2,9,1,0,1,4,1,3,1,1,-3,-5,4]
```

```python
# Copy of original data
temps_r = temps.copy()
```

```python
# Data put in chronological order
temps.reverse()
```

```python
# Average temperature difference for whole time period
sum(temps)/len(temps)
```

```python
# First part of list removed (since trend most clearly began in late January) and average recalculated
temps_plot = temps[26:]
avg = round(sum(temps_plot)/len(temps_plot),1)
avg
```

```python
# Values and indexes for above average temperatures (for plotting purposes)
above_avg = [num for num in temps_plot if num>=0]
```

```python
above_avg_ind = [i for i,v in enumerate(temps_plot) if v>=0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data Visualization

```python
plt.figure(figsize=(12,5))
plt.plot(temps_plot)
plt.scatter(x=above_avg_ind,y=above_avg,color='red')
plt.title('Oakland 2023 Temperatures vs. Historical Average')
plt.ylabel('Degrees F above/below historical average')
plt.xlabel('Date')
plt.xticks([5,33,64,94,125,155],labels=['February','March','April','May','June','July'])
plt.text(-3,8, f'{len(above_avg)} days ({round(len(above_avg)*100/len(temps_plot))}%) at or above the historical average for that date \n\n{len(temps_plot)-len(above_avg)} days ({round((len(temps_plot)-len(above_avg))*100/len(temps_plot))}%) below the historical average for that date',color='red')
plt.text(-6,-7.5,f'Overall average: \n            {avg} deg.',color='red')
plt.axhline(0,color='black')
plt.axhline(avg,color='red',ls='--',lw=0.5)
```
