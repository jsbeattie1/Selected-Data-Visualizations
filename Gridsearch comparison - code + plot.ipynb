## Summary/background

**Comparison of 4 different gridsearch options, with the resulting model accuracies and gridsearch processing times shown on a combined bar/line graph.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
models = ['GridSearchCV','HalvingGridSearchCV','HalvingRandomSearchCV','RandomizedSearchCV']
accuracies = [92.10,91.35,92.86,92.76]
times = [24.86,21.15,5.44,0.90]

# Data from gridsearches carried out in another notebook
```

```python
fig, ax1 = plt.subplots(figsize = (12,5))
ax2 = plt.twinx(ax = ax1)
plt.suptitle('Accuracy and Processing Time for Different Grid Search Methods',weight='bold')
sns.barplot(x=models,y=accuracies,ax=ax1,color='gray',alpha=0.2)
sns.lineplot(x=models,y=times,ax=ax2,linestyle='dotted',color='red')
sns.scatterplot(x=models,y=times,ax=ax2,color='red')
ax1.set_ylim(0,100)
ax1.set_ylabel('Accuracy (%)',fontsize=12)
ax2.set_ylabel('Processing Time (seconds)',fontsize=12,color='red')
ax2.set_yticks([0,5,10,15,20,25,30,35])
ax2.set_yticklabels([0,5,10,15,20,25,30,35],color='red')
for x,y in zip(models,times):
    plt.text(x,y+1.1,'{:.1f} s'.format(y),color='red',ha='center',size='11')
for x,y in zip(models,accuracies):
    plt.text(x,30.3,'{:.1f}%'.format(y),color='black',ha='center',weight='bold',size='11')
```
